# Titanic passenger survival

A look at the classic Titanic passenger dataset (891 passengers): text stats with Python's standard library, then charts with matplotlib.

```python
import csv
import kernel_runner

with open(kernel_runner.data_path("titanic.csv")) as f:
    rows = list(csv.DictReader(f))

print(f"Loaded {len(rows)} passengers")
```

## Overall survival rate

```python
survived = sum(1 for r in rows if r["Survived"] == "1")
rate = survived / len(rows) * 100
print(f"Overall survival rate: {rate:.1f}% ({survived}/{len(rows)})")
```

## By sex

```python
from collections import defaultdict

def survival_counts(key):
    groups = defaultdict(lambda: [0, 0])
    for r in rows:
        g = groups[r[key]]
        g[1] += 1
        if r["Survived"] == "1":
            g[0] += 1
    return groups

def survival_rate_by(key):
    for value, (s, t) in sorted(survival_counts(key).items()):
        print(f"{value}: {s}/{t} ({s / t * 100:.1f}%)")

survival_rate_by("Sex")
```

## By passenger class

```python
survival_rate_by("Pclass")
```

## Age

```python
ages_survived = [float(r["Age"]) for r in rows if r["Survived"] == "1" and r["Age"]]
ages_died = [float(r["Age"]) for r in rows if r["Survived"] == "0" and r["Age"]]

print(f"Average age (survived): {sum(ages_survived) / len(ages_survived):.1f}")
print(f"Average age (did not survive): {sum(ages_died) / len(ages_died):.1f}")
```

## Visualizations

The cells below chart the same findings with the bundled **matplotlib**. Run the notebook top-to-bottom — they reuse `rows` loaded in the first cell.

```python
import matplotlib.pyplot as plt

def rates(key):
    groups = survival_counts(key)
    labels = sorted(groups)
    return labels, [groups[l][0] / groups[l][1] * 100 for l in labels]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

sx, sy = rates("Sex")
ax1.bar(sx, sy, color=["#4C72B0", "#DD8452"])
ax1.set_title("Survival rate by sex")
ax1.set_ylabel("% survived")

cx, cy = rates("Pclass")
ax2.bar([f"Class {c}" for c in cx], cy, color="#55A868")
ax2.set_title("Survival rate by class")

fig.tight_layout()
```

## Age distribution

```python
plt.figure(figsize=(8, 4))
plt.hist(
    [ages_survived, ages_died],
    bins=16,
    label=["Survived", "Did not survive"],
    color=["#55A868", "#C44E52"],
)
plt.xlabel("Age")
plt.ylabel("Passengers")
plt.title("Age distribution by survival")
plt.legend()
plt.show()
```
