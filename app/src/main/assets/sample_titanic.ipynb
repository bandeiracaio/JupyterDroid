# Titanic passenger survival

A quick text-based look at the classic Titanic passenger dataset (891 passengers) using only Python's standard library — no external packages required.

```python
import csv
import kernel_runner

with open(kernel_runner.data_path("titanic.csv")) as f:
    rows = list(csv.DictReader(f))

print(f"Loaded {len(rows)} passengers")
```

## Overall survival rate

```python
survived = sum(1 for r in rows if r["Survived"] == "1")
rate = survived / len(rows) * 100
print(f"Overall survival rate: {rate:.1f}% ({survived}/{len(rows)})")
```

## By sex

```python
from collections import defaultdict

def survival_rate_by(key):
    groups = defaultdict(lambda: [0, 0])
    for r in rows:
        g = groups[r[key]]
        g[1] += 1
        if r["Survived"] == "1":
            g[0] += 1
    for value, (s, t) in sorted(groups.items()):
        print(f"{value}: {s}/{t} ({s / t * 100:.1f}%)")

survival_rate_by("Sex")
```

## By passenger class

```python
survival_rate_by("Pclass")
```

## Age

```python
ages_survived = [float(r["Age"]) for r in rows if r["Survived"] == "1" and r["Age"]]
ages_died = [float(r["Age"]) for r in rows if r["Survived"] == "0" and r["Age"]]

print(f"Average age (survived): {sum(ages_survived) / len(ages_survived):.1f}")
print(f"Average age (did not survive): {sum(ages_died) / len(ages_died):.1f}")
```
